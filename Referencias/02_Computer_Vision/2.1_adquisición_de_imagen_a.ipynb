# ADQUISICION DE IMAGEN Y ESPACIOS DE COLOR

# Se cargan las bibliotecas necesarias

```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
```

# Se lee la imagen y se imprimen sus dimensiones

```python
img = cv2.imread('Self_Driving_Car.jpg')
```

```python
img.shape
print ('Altura = ', int(img.shape[0]), 'pixels')
print ('Anchura = ', int(img.shape[1]), 'pixels')
```

# Se despliega la imagen con OpenCV en una ventana independiente

```python
# Al ejecutar esta celda de código, se crea una ventana independiente
# waitKey() espera a que se presione cualquier tecla del teclado
# Después de que se presiona una tecla, destroyAllWindows() cierra la ventana de la imagen

cv2.imshow('Self Driving Car!', img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Alternativamente, se puede usar Matplotlib para mostrar la imagen en línea con el código

```python
#nótese que existe un cambio en el orden de los planos de color
plt.imshow(img)
#se imprimen de nuevo las dimensiones de la imagen
img.shape
```

# Se convierte la imagen a escala de grises

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow('Self Driving Car in Grayscale!', gray_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Se imprimen las dimensiones de la imagen en escala de grises

```python
gray_img.shape
```

# Se convierte la imagen a color en un arreglo bidimensional usando Numpy y se usa Matplotlib para mostrarla

```python
#nuevamente se presenta el cambio en el orden de los planos de color
img_arr = np.asarray(img)
img_arr.shape
plt.imshow(img_arr)
```

# Se usa la biblioteca PIL para leer la misma imagen de color y se muestra

```python
img2 = Image.open('Self_Driving_Car.jpg')
```

# La imagen se convierte en arreglo y se muestra usando Matplotlib

```python
#nótese que en esta ocasión no hay error en los colores
img_arr = np.asarray(img2)
img_arr.shape
plt.imshow(img_arr)
```

# Se crea una copia del arreglo y se muestra el plano rojo

```python
img_rojo = img_arr.copy()
#el plano verde se hace cero
img_rojo[:, :, 1] = 0 
#el plano azul se hace cer0
img_rojo[:, :, 2] = 0 
plt.imshow(img_rojo)
```

# Se convierte la imagen de color al espacio HSV

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV Image', hsv_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Se muestra el plano de *Hue*

```python
plt.imshow(hsv_img[:, :, 0])
plt.title('Hue')
```

# Se muestra el plano de *Saturation*

```python
plt.imshow(hsv_img[:, :, 1])
plt.title('Saturation')
```

# Se muestra el canal de *Value*

```python
plt.imshow(hsv_img[:, :, 2])
plt.title('Value')
```
